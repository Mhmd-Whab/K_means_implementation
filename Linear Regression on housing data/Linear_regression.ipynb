```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import load_boston

from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, Normalizer, RobustScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV

from sklearn.pipeline import make_pipeline

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
from sklearn import metrics

import warnings
warnings.filterwarnings('ignore')
```

```python
data = load_boston()
X = data.data
Y = data.target
X.shape, Y.shape
```

```python
df_x = pd.DataFrame(X)
df_y = pd.DataFrame(Y)
df_x
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.1, random_state=101)
x_train.shape, x_test.shape
```

```python
def polynomial_features(degree, train, test):
    
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    train_poly = poly.fit_transform(train)
    test_poly = poly.transform(test)
    
    return train_poly, test_poly
```

```python
def polynomial_regression(degree, cv, train, test):
    
    train_poly, test_poly = polynomial_features(degree, train, test)
    
    
    print('Train data and test data shapes are: ', train_poly.shape, test_poly.shape)
    
    model = LinearRegression(normalize=True)
    param_grid = {'fit_intercept': [True]}

    print('----------------------------------')
    print('degree = ', degree)
    print('CV = ', cv)
    print('----------------------------------')
    
    model_gscv = GridSearchCV(model, param_grid = param_grid, cv=cv, 
                              verbose=3, scoring='neg_mean_absolute_error')
    model_gscv.fit(train_poly, y_train)
    
    print('----------------------------------')
    scores = []
    best_cv = 0
    for i in range(cv):

        scores.append(model_gscv.cv_results_['split'+str(i)+'_test_score'][0])
        best_cv = scores.index(max(scores))
    best_score = -max(scores)
    print('mean_score = ', -np.mean(scores))
    print('best_score = ', best_score)
    print('best_cv = ', best_cv+1)
    print('----------------------------------')
    
    print('*******************************************')
    return 
```

```python
for i in range(1,11):
    
    polynomial_regression(i, 5, x_train, x_test)
```

#### Ridge regression

#### alpha: must be positive, increase for more regularization

```python
def range_with_floats(start, stop, step):
    while stop > start:
        yield start
        start += step
```

```python
for a in range_with_floats(0, 0.1, 0.01):
    
    ridgereg = Ridge(alpha=a, normalize=True)
    ridgereg.fit(x_train_poly_slice, y_train_slice)

    y_pred_train = ridgereg.predict(x_train_poly_slice)
    y_pred_test = ridgereg.predict(x_test_poly)
    
    print('alpha = ', a)
    print('-----------------------')
    print('train data score = ', mean_absolute_error(y_pred_train, y_train_slice))
    print('test data score = ', mean_absolute_error(y_pred_test, y_test))
    print('-----------------------')
    
    print("R-Square Value",r2_score(y_test,y_pred_test))
    print("\n")
    print ("mean_squared_error : ",metrics.mean_squared_error(y_test, y_pred_test))
    print("\n")
    print ("root_mean_squared_error : ",np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)))
    print("\n")
    
    print('-----------------------')
    print('*******************************************')

# print ("iNTERCEPT : ",ridgereg.intercept_)
# print ("CO-EFFICIENT : ",ridgereg.coef_)
```

```python
# create an array of alpha values
alpha_range = [*range_with_floats(0.01, 0.1, 0.01)]

# select the best alpha with RidgeCV
ridgeregcv = RidgeCV(alphas=alpha_range, normalize=True, scoring='neg_mean_absolute_error')
ridgeregcv.fit(x_train_poly_slice, y_train_slice)
print('ridgeregcv.alpha = ' , ridgeregcv.alpha_)


# predict method uses the best alpha value
y_pred_train = ridgeregcv.predict(x_train_poly_slice)
y_pred_test = ridgeregcv.predict(x_test_poly)
# calculate R^2 value, MAE, MSE, RMSE

print('-----------------------')
print('train data score = ', mean_absolute_error(y_pred_train, y_train_slice))
print('test data score = ', mean_absolute_error(y_pred_test, y_test))
print('-----------------------')

print("R-Square Value",r2_score(y_test,y_pred_test))
print("\n")
print ("mean_absolute_error :",metrics.mean_absolute_error(y_test, y_pred_test))
print("\n")
print ("mean_squared_error : ",metrics.mean_squared_error(y_test, y_pred_test))
print("\n")
print ("root_mean_squared_error : ",np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)))
```
